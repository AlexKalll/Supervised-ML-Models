## 3. Mathematical Foundation

In ML, data is usually stored as numbers or text in tables. Here are some basic math ideas you will use in Python:

- Scalars: a single number, like 5 or 3.14.
- Vectors: a list of numbers, like [1, 2, 3].
- Matrices: a table of numbers with rows and columns.

Key operations:
- Addition: element-wise addition of numbers or arrays.
- Mean: the average of values, computed as sum(values) / count(values).
- Standard deviation: how spread out values are around the mean.

Example formula for mean:

mean = (x_1 + x_2 + ... + x_n) / n

Where:
- x_i is each value in the list
- n is the number of values

These simple operations are the building blocks for cleaning and exploring data.

```python
# Import the Python libraries we will use in this notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Set a visual style for plots
sns.set(style="whitegrid")

print("Libraries imported successfully.")
```

## 4. Implementation (Python)

### 4.1 Numpy Basics
Numpy is the core library for numerical computing. It stores numbers efficiently in arrays and allows fast operations on those arrays.

```python
# Create a one-dimensional array (vector)
vector = np.array([1, 2, 3, 4, 5])
print('Vector:', vector)

# Create a two-dimensional array (matrix)
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print('Matrix:
', matrix)

# Compute the mean and standard deviation
mean_value = vector.mean()
std_value = vector.std()
print('Mean:', mean_value)
print('Standard deviation:', std_value)
```

### 4.2 Numpy Operations
We can perform arithmetic on arrays element by element. This is useful for transforming data quickly.

```python
# Element-wise operations with numpy arrays
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print('a + b =', a + b)
print('a - b =', a - b)
print('a * b =', a * b)
print('a / b =', a / b)

# Reshape an array into a matrix with 3 rows and 1 column
reshaped = a.reshape(3, 1)
print('Reshaped array:', reshaped)
```

### 4.3 Pandas Basics
Pandas is used to work with labeled data like tables. The main structures are Series (one column) and DataFrame (multiple columns).

```python
# Create a pandas Series from a list
series = pd.Series([10, 20, 30, 40], name='Sales')
print('Series:
', series)

# Create a pandas DataFrame from a dictionary
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
    'Sales': [10, 15, 13, 20],
    'Expenses': [8, 12, 11, 18]
}
df = pd.DataFrame(data)
print('
DataFrame:
', df)

# Show the first rows of the DataFrame
print('
First rows:
', df.head())

# Describe the numerical columns
print('
Statistics:
', df.describe())
```

### 4.4 Data Selection with Pandas
Selecting rows and columns lets you focus on the most important data for analysis.

```python
# Select the 'Sales' column
sales_column = df['Sales']
print('Sales column:
', sales_column)

# Select the row for March
march_row = df.loc[2]  # index 2 is March
print('
March row:
', march_row)

# Filter rows where Sales are greater than 12
high_sales = df[df['Sales'] > 12]
print('
High sales rows:
', high_sales)
```

## 5. Visualization
Visualization helps you see patterns, trends, and relationships in your data. We use Matplotlib and Seaborn to draw charts.

```python
# Plot Sales and Expenses over months
plt.figure(figsize=(8, 5))
plt.plot(df['Month'], df['Sales'], marker='o', label='Sales')
plt.plot(df['Month'], df['Expenses'], marker='o', label='Expenses')
plt.title('Sales and Expenses by Month')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a bar plot using seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x='Month', y='Sales', data=df, palette='Blues_d')
plt.title('Monthly Sales')
plt.ylabel('Sales')
plt.show()
```

## 7. Mini Project
Use a small dataset to practice Python for ML. This mini project uses the Iris dataset, a classic dataset in machine learning.

### Goal
Load the Iris dataset, inspect the data, and visualize the flower measurements.

```python
# Load the Iris dataset from scikit-learn
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Show the first 5 rows of the dataset
print(iris_df.head())

# Compute basic statistics
print('
Iris dataset statistics:
', iris_df.describe())

# Visualize the petal length distribution by species
plt.figure(figsize=(8, 5))
sns.boxplot(x='species', y='petal length (cm)', data=iris_df, palette='Set2')
plt.title('Petal Length Distribution by Species')
plt.show()
```
