# Data Preprocessing for Machine Learning

This notebook explains how to clean, transform, and prepare data for machine learning models. Data preprocessing is the second essential step in every ML workflow.

## 1. Introduction

### What is the topic?
Data preprocessing is the process of cleaning and converting raw data into a format that machine learning algorithms can understand.

### Why it matters in ML
Raw data is often messy, incomplete, or inconsistent. Preprocessing helps improve model accuracy, reduce errors, and make ML training faster and more reliable.

## 2. Intuition First

Think of preprocessing like washing and chopping vegetables before cooking. If you use dirty or uneven ingredients, the final meal may not taste good. In ML, models perform better when the data is clean and consistent.

Real-life examples:
- Removing empty fields is like throwing away rotten fruit.
- Converting text labels into numbers is like translating a recipe into a language the computer can follow.
- Scaling values is like using the same measurement units for all ingredients.

## 3. Mathematical Foundation

Key preprocessing operations are based on simple math and transformations:

1. Missing value replacement:
   - mean imputation: x_i = mean(x) for missing entries

2. Standardization (z-score):
   - z = (x - mu) / sigma
   - where x is a data point, mu is the mean, and sigma is the standard deviation

3. Min-max scaling:
   - x_scaled = (x - min) / (max - min)
   - this maps values into the range [0, 1]

4. One-hot encoding for categorical values:
   - convert categories like 'red', 'green', 'blue' into separate binary features

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

sns.set(style="whitegrid")
print('Libraries loaded successfully.')
```

## 4. Implementation (Python)

### 4.1 Create a sample dataset
We will use a small example dataset to demonstrate cleaning, handling missing values, scaling, and encoding.

```python
data = {
    'Age': [25, 30, np.nan, 22, 40],
    'Salary': [50000, 60000, 52000, np.nan, 58000],
    'Department': ['Sales', 'Engineering', 'Sales', 'HR', 'Engineering'],
    'Purchased': ['No', 'Yes', 'No', 'Yes', 'No']
}
df = pd.DataFrame(data)
print('Original data:')
print(df)
```

### 4.2 Inspect the data
We check the shape, missing values, and data types to understand what needs cleaning.

```python
print('Data types:')
print(df.dtypes)

print('
Missing values:')
print(df.isnull().sum())

print('
Summary statistics:')
print(df.describe(include='all'))
```

### 4.3 Handle missing values
We replace missing numerical values with the mean using SimpleImputer.

```python
num_imputer = SimpleImputer(strategy='mean')
numeric_columns = ['Age', 'Salary']

# Fit the imputer only on numeric columns
df[numeric_columns] = num_imputer.fit_transform(df[numeric_columns])
print('Data after imputing missing values:')
print(df)
```

### 4.4 Scale numeric features
Scaling puts numeric columns on the same scale. We use both StandardScaler and MinMaxScaler to compare.

```python
scaler = StandardScaler()
scaled_values = scaler.fit_transform(df[numeric_columns])
scaled_df = pd.DataFrame(scaled_values, columns=['Age_scaled', 'Salary_scaled'])
print('Standard-scaled numeric data:')
print(scaled_df)

minmax_scaler = MinMaxScaler()
minmax_values = minmax_scaler.fit_transform(df[numeric_columns])
minmax_df = pd.DataFrame(minmax_values, columns=['Age_minmax', 'Salary_minmax'])
print('
Min-max scaled numeric data:')
print(minmax_df)
```

### 4.5 Encode categorical variables
Machine learning models require numeric input, so we transform categories into numbers.

```python
encoder = OneHotEncoder(sparse=False, drop='first')
cat_columns = ['Department']
encoded_values = encoder.fit_transform(df[cat_columns])
encoded_df = pd.DataFrame(encoded_values, columns=encoder.get_feature_names_out(cat_columns))
print('One-hot encoded categorical data:')
print(encoded_df)
```

## 5. Visualization
We visualize the distribution of numeric values before and after scaling to see how preprocessing changes the data.

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde=True, color='blue')
plt.title('Age after Imputation')

plt.subplot(1, 2, 2)
sns.histplot(scaled_df['Age_scaled'], kde=True, color='green')
plt.title('Age after Standard Scaling')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=encoded_df.columns, y=encoded_df.iloc[0], palette='pastel')
plt.title('Example One-Hot Encoded Row')
plt.ylabel('Value')
plt.show()
```

### 5.1 Preprocessing pipeline
A pipeline automates preprocessing steps so the same transformations are applied consistently.

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])
categorical_transformer = Pipeline(steps=[
    ('encoder', OneHotEncoder(drop='first', sparse=False))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_columns),
    ('cat', categorical_transformer, cat_columns)
])

processed_array = preprocessor.fit_transform(df)
processed_df = pd.DataFrame(processed_array, columns=['Age_scaled', 'Salary_scaled'] + list(encoder.get_feature_names_out(cat_columns)))
print('Data after preprocessing pipeline:')
print(processed_df)
```

## 7. Mini Project
Use a small dataset to preprocess features before model training.

### Goal
Load the Titanic dataset, fill missing values, encode categorical columns, and scale numeric features.

Note: This is a preprocessing practice step, not model training yet.

```python
titanic_data = {
    'Age': [22, 38, 26, 35, np.nan],
    'Fare': [7.25, 71.2833, 7.925, 53.1, 8.05],
    'Sex': ['male', 'female', 'female', 'female', 'male'],
    'Embarked': ['S', 'C', 'S', 'S', 'S']
}
titanic_df = pd.DataFrame(titanic_data)
print('Titanic sample data:')
print(titanic_df)

# Fill missing Age values with the mean
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())

# Encode categorical variables using one-hot encoding
titanic_df = pd.get_dummies(titanic_df, columns=['Sex', 'Embarked'], drop_first=True)

# Scale numeric columns
scaler = MinMaxScaler()
numeric_cols = ['Age', 'Fare']
titanic_df[numeric_cols] = scaler.fit_transform(titanic_df[numeric_cols])

print('
Preprocessed Titanic data:')
print(titanic_df)
```

## 8. Summary
- Data preprocessing cleans and prepares raw data for machine learning.
- Missing values can be filled using the mean, median, or mode.
- Scaling and standardization help numeric features work better together.
- Categorical values must be converted into numbers before training models.
- Pipelines help keep preprocessing consistent and reproducible.
